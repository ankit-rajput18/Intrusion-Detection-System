# TON_IoT Inference Notebook

## 11) Create a Separate Inference `.ipynb` Using Saved Artifacts
This notebook reuses exported artifacts from training to predict classes and probabilities on new TON_IoT samples.

```python
import json
import numpy as np
import pandas as pd
import joblib

ARTIFACT_DIR = '../models/ton_iot'
DATA_PATH = '../data/ton_iot/train_test_network.csv'

with open(f'{ARTIFACT_DIR}/ton_iot_metadata.json', 'r', encoding='utf-8') as f:
    metadata = json.load(f)

model = joblib.load(f"{ARTIFACT_DIR}/ton_iot_model_pipeline.joblib")
label_encoder = joblib.load(f"{ARTIFACT_DIR}/ton_iot_label_encoder.joblib")

feature_columns = metadata['feature_columns']
print('Loaded artifacts successfully.')
print('Target column used in training:', metadata['target_column'])
print('Number of classes:', len(metadata['classes']))
```

```python
df_new = pd.read_csv(DATA_PATH)

# Match training-time placeholder cleaning
for token in ['-', ' -', '- ', 'None', 'none', 'N/A', 'na', '']:
    df_new.replace(token, np.nan, inplace=True)

for col in df_new.columns:
    if col in ['label', 'type']:
        continue
    as_num = pd.to_numeric(df_new[col], errors='coerce')
    if as_num.notna().mean() >= 0.70:
        df_new[col] = as_num
    else:
        df_new[col] = df_new[col].astype(str).str.strip().str.lower().replace('nan', np.nan)

for col in ['src_ip', 'dst_ip']:
    if col in df_new.columns and col not in feature_columns:
        df_new = df_new.drop(columns=[col])

X_new = df_new.reindex(columns=feature_columns)
X_new = X_new.head(20)
print('Inference batch shape:', X_new.shape)
```

```python
pred_enc = model.predict(X_new)
pred_label = label_encoder.inverse_transform(pred_enc)

if hasattr(model, 'predict_proba'):
    pred_proba = model.predict_proba(X_new)
    confidence = pred_proba.max(axis=1)
else:
    pred_proba = None
    confidence = np.full(len(pred_label), np.nan)

out = pd.DataFrame({
    'predicted_class': pred_label,
    'confidence': confidence,
})

display(out.head(20))
```

```python
out.to_csv(f'{ARTIFACT_DIR}/ton_iot_sample_predictions.csv', index=False)
print('Saved sample predictions to models/ton_iot/ton_iot_sample_predictions.csv')
```
