Import Libraries

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from imblearn.over_sampling import SMOTE
```

Load Dataset

```python
df = pd.read_csv('../data/train.csv')
print(df.shape)
df.head()
```

```python
print(df['label'].value_counts())
```

```python
# Encode labels for Stage 1
stage1_labels = ['Benign', 'DDoS', 'DoS', 'Malware', 'Other', 'Recon', 'WebAttack']  # define Stage 1 classes
df['stage1_label'] = df['label'].apply(lambda x: x if x in stage1_labels else 'Other')

# Features and labels for Stage 1
X = df.drop(['label', 'stage1_label'], axis=1)
y_stage1 = df['stage1_label']

# Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_stage1, test_size=0.2, random_state=42, stratify=y_stage1)
```

Stage 1 Label Mapping (Broad Classification)

```python
# Random Forest for Stage 1
stage1_model = RandomForestClassifier(n_estimators=100, max_depth=20, class_weight='balanced', n_jobs=-1, random_state=42)
stage1_model.fit(X_train, y_train)

# Stage 1 predictions
y_pred_stage1 = stage1_model.predict(X_test)

# Evaluate Stage 1
print("Stage 1 Accuracy:", accuracy_score(y_test, y_pred_stage1))
print("Stage 1 Confusion Matrix:\n", confusion_matrix(y_test, y_pred_stage1, labels=np.unique(y_test)))
print("\nStage 1 Classification Report:\n", classification_report(y_test, y_pred_stage1))
```

Prepare Features and Labels for Stage 1

```python
rare_classes = ['WebAttack', 'Recon', 'MalwareOther']  # rare classes to separate
stage2_mask = df['label'].isin(rare_classes)
df_stage2 = df[stage2_mask]

X2 = df_stage2.drop(['label', 'stage1_label'], axis=1)
y2 = df_stage2['label']
```

```python
# Stage 2 Random Forest
stage2_model = RandomForestClassifier(n_estimators=100, max_depth=20, class_weight='balanced', n_jobs=-1, random_state=42)
stage2_model.fit(X2_train, y2_train)
```

```python
final_pred = []

for i, pred in enumerate(y_pred_stage1):
    if pred == 'Other':
        # Only predict with Stage 2 if there are enough samples
        if len(np.unique(y2_train)) > 1:
            stage2_pred = stage2_model.predict([X_test[i]])[0]
            final_pred.append(le_stage2.inverse_transform([stage2_pred])[0])
        else:
            final_pred.append('Other')
    else:
        final_pred.append(pred)

# True labels
y_true = df.loc[X_test.index, 'label']
```

```python
# Accuracy
acc = accuracy_score(y_true, final_pred)
print("Full Pipeline Accuracy:", acc)

# Confusion Matrix
cm = confusion_matrix(y_true, final_pred, labels=np.unique(y_true))
print("Confusion Matrix:\n", cm)

# Classification Report
print("\nClassification Report:\n", classification_report(y_true, final_pred))
```
