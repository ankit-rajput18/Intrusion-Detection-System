# Two-Stage Intrusion Detection System (IDS)
This notebook implements a **two-stage Random Forest IDS** with separate train and test CSV files.  
- **Stage 1**: Coarse classification (DDoS, DoS, Recon, Malware, WebAttack, Benign, Other)  
- **Stage 2**: Fine-grained classification of rare/Other classes using SMOTE balancing  

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, f1_score, balanced_accuracy_score
from imblearn.over_sampling import SMOTE
```

```python
import warnings
warnings.filterwarnings("ignore")
```

## Load Train and Test Datasets
There is a **train.csv** and  a **test.csv** exist in the working directory.

```python
df_train = pd.read_csv("../data/train.csv")
df_test = pd.read_csv("../data/test.csv")
```

## Stage 1 Label Mapping (Aligned Label Space)
Stage 1 predicts **frequent fine-grained labels** directly and maps rare classes to **'Other'**.
This ensures Stage 1 output can be merged with Stage 2 output into the same final fine-label space.

```python
# Build Stage 1 labels: frequent fine labels + 'Other'
min_count_stage1 = 500
label_counts = df_train['label'].value_counts()
frequent_labels = set(label_counts[label_counts >= min_count_stage1].index)

def stage1_label(label):
    return label if label in frequent_labels else 'Other'

df_train['stage1_label'] = df_train['label'].apply(stage1_label)
df_test['stage1_label'] = df_test['label'].apply(stage1_label)

print('Frequent Stage 1 labels:', sorted(frequent_labels))
print(df_train['stage1_label'].value_counts())
print('Train Other count:', int((df_train['stage1_label'] == 'Other').sum()))
print('Test Other count:', int((df_test['stage1_label'] == 'Other').sum()))
```

## Stage 1 Features & Targets
Prepare feature matrices plus both evaluation targets:
- `y_train_stage1`: frequent fine labels + 'Other' (for Stage 1 training)
- `y_test_stage1`: same mapped coarse/frequent space (for Stage 1 evaluation)
- `y_test_fine`: original fine labels (for end-to-end evaluation)

```python
# Sample data to fit in memory (use 500K train, 100K test)
TRAIN_SAMPLE_SIZE = 500000
TEST_SAMPLE_SIZE = 100000

train_indices = np.random.RandomState(42).choice(len(df_train), size=min(TRAIN_SAMPLE_SIZE, len(df_train)), replace=False)
test_indices = np.random.RandomState(42).choice(len(df_test), size=min(TEST_SAMPLE_SIZE, len(df_test)), replace=False)

X_train = df_train.iloc[train_indices].drop(columns=['label', 'stage1_label']).values.astype(np.float32)
X_test = df_test.iloc[test_indices].drop(columns=['label', 'stage1_label']).values.astype(np.float32)

y_train_stage1 = df_train.iloc[train_indices]['stage1_label'].values
y_test_stage1 = df_test.iloc[test_indices]['stage1_label'].values
y_test_fine = df_test.iloc[test_indices]['label'].values

print('Sampled train/test for memory efficiency')
print('X_train shape:', X_train.shape, 'X_test shape:', X_test.shape)
```

## Train Stage 1 Random Forest Model
We use class_weight='balanced' to account for imbalance in coarse categories.

```python
stage1_model = RandomForestClassifier(
    n_estimators=100, max_depth=15, class_weight='balanced', n_jobs=1, random_state=42
)
stage1_model.fit(X_train, y_train_stage1)
```

## Stage 2: Fine-Grained Classification for `Other`
- Train only on samples where Stage 1 target is `Other`.
- Predict original fine-grained rare labels.
- Use SMOTE only when valid classes have enough samples.

```python
# Stage 2: prepare data from train set
train_other_mask = (y_train_stage1 == 'Other')
X2 = X_train[train_other_mask]
y2 = df_train.iloc[train_indices[train_other_mask]]['label'].values

# Keep only classes with enough samples for SMOTE
min_samples_stage2 = 2
y2_counts = pd.Series(y2).value_counts()
valid_classes = y2_counts[y2_counts >= min_samples_stage2].index.values

X2 = X2[np.isin(y2, valid_classes)]
y2 = y2[np.isin(y2, valid_classes)]

stage2_enabled = len(y2) > 0 and len(np.unique(y2)) >= 2
le_stage2 = None
X2_res, y2_res = None, None

if not stage2_enabled:
    print('Stage 2 disabled: no valid training samples after filtering.')
else:
    # Encode Stage 2 labels
    le_stage2 = LabelEncoder()
    y2_enc = le_stage2.fit_transform(y2)

    # Adapt k_neighbors to smallest class size
    min_class_size = pd.Series(y2).value_counts().min()
    k_neighbors = max(1, min(5, min_class_size - 1))
    smote = SMOTE(random_state=42, k_neighbors=k_neighbors)
    X2_res, y2_res = smote.fit_resample(X2, y2_enc)

    print('Stage 2 classes:', list(le_stage2.classes_))
    print('Stage 2 class counts before SMOTE:', pd.Series(y2).value_counts().to_dict())
    print('Stage 2 resampled shape:', X2_res.shape)
```

## Train Stage 2 Random Forest Model
Only trains on balanced 'Other' category.

```python
if stage2_enabled:
    stage2_model = RandomForestClassifier(
        n_estimators=100, max_depth=15, class_weight='balanced', n_jobs=1, random_state=42
    )
    stage2_model.fit(X2_res, y2_res)
else:
    stage2_model = None
    print('Skipped Stage 2 model training.')
```

## Prediction Pipeline
- Stage 1 predicts frequent fine labels or `Other`.
- Only `Other` predictions are sent to Stage 2.
- Final output is always in the original fine-label space.

```python
y_pred_stage1 = stage1_model.predict(X_test)

# Final predictions start in Stage 1 space (frequent fine labels + 'Other')
final_pred = y_pred_stage1.copy()

# Route only Stage-1 'Other' predictions to Stage 2 when Stage 2 is available
other_indices = np.where(y_pred_stage1 == 'Other')[0]

if len(other_indices) > 0 and stage2_enabled and stage2_model is not None:
    X_stage2 = X_test[other_indices]
    stage2_preds = stage2_model.predict(X_stage2)
    stage2_preds = le_stage2.inverse_transform(stage2_preds)
    final_pred[other_indices] = stage2_preds

final_pred = np.array(final_pred, dtype=object)
```

## Evaluation (Properly Aligned)
- **Stage 1 Evaluation** in Stage 1 label space (`y_test_stage1` vs `y_pred_stage1`).
- **Stage 2 Evaluation** on true `Other` subset (optional diagnostic).
- **End-to-End Evaluation** in original fine-label space (`y_test_fine` vs `final_pred`).

```python
# 1) Stage 1 evaluation in Stage 1 label space
stage1_acc = accuracy_score(y_test_stage1, y_pred_stage1)
stage1_macro_f1 = f1_score(y_test_stage1, y_pred_stage1, average='macro', zero_division=0)
stage1_bal_acc = balanced_accuracy_score(y_test_stage1, y_pred_stage1)

print('Stage 1 Accuracy (mapped space):', stage1_acc)
print('Stage 1 Macro-F1 (mapped space):', stage1_macro_f1)
print('Stage 1 Balanced Accuracy (mapped space):', stage1_bal_acc)
print('\nStage 1 Classification Report:\n', classification_report(y_test_stage1, y_pred_stage1, zero_division=0))

# 2) Stage 2 diagnostic evaluation on true-Other subset (if Stage 2 exists)
if stage2_enabled:
    mask_true_other = (y_test_stage1 == 'Other')
    if mask_true_other.sum() > 0:
        y2_true_test = y_test_fine[mask_true_other]
        y2_pred_test = final_pred[mask_true_other]
        print('\nStage 2 Diagnostic Report (true Other subset):\n',
              classification_report(y2_true_test, y2_pred_test, zero_division=0))
    else:
        print('\nNo true Other samples in test set for Stage 2 diagnostic evaluation.')
else:
    print('\nStage 2 Diagnostic Report: skipped because Stage 2 is disabled.')

# 3) End-to-end evaluation in original fine-label space
pipeline_acc = accuracy_score(y_test_fine, final_pred)
pipeline_macro_f1 = f1_score(y_test_fine, final_pred, average='macro', zero_division=0)
pipeline_weighted_f1 = f1_score(y_test_fine, final_pred, average='weighted', zero_division=0)
all_labels = np.unique(np.concatenate([y_test_fine, final_pred]))
cm = confusion_matrix(y_test_fine, final_pred, labels=all_labels)
report = classification_report(y_test_fine, final_pred, zero_division=0)

print('\nEnd-to-End Accuracy (fine-label space):', pipeline_acc)
print('End-to-End Macro-F1 (fine-label space):', pipeline_macro_f1)
print('End-to-End Weighted-F1 (fine-label space):', pipeline_weighted_f1)

# Benign false positive rate (if Benign exists in labels)
if 'Benign' in all_labels:
    benign_mask = (y_test_fine == 'Benign')
    if benign_mask.sum() > 0:
        benign_fp_rate = np.mean(final_pred[benign_mask] != 'Benign')
        print('Benign False Positive Rate:', benign_fp_rate)

print('\nConfusion Matrix (fine-label space):\n', cm)
print('\nEnd-to-End Classification Report:\n', report)
```

```python
# Class count summary
print('=== CLASS BREAKDOWN ===')
print('Stage 1 (mapped) classes:', len(np.unique(y_train_stage1)))
print('Classes:', sorted(np.unique(y_train_stage1)))
print()
print('Original fine-grained classes:', len(np.unique(y_test_fine)))
print('Class distribution in test:', pd.Series(y_test_fine).value_counts().to_dict())
```

## Ablation Study (Paper-Ready)

This section compares three variants on the **same sampled split**:
1. **Stage1 Only**
2. **Stage1 + Stage2 (Default SMOTE)**
3. **Stage1 + Stage2 (Tuned SMOTE)**

Primary metrics reported:
- Pipeline Accuracy
- Pipeline Macro-F1
- Pipeline Weighted-F1
- Benign False Positive Rate
- Rare-Class Recall (classes with test count < `rare_threshold`)

```python
from sklearn.metrics import accuracy_score, f1_score, classification_report
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
from imblearn.over_sampling import SMOTE

# Optional: adjust this threshold for what counts as "rare"
rare_threshold = 500

# Train one Stage 1 model (same for all ablations)
stage1_model_ablation = RandomForestClassifier(
    n_estimators=100,
    max_depth=15,
    class_weight='balanced',
    n_jobs=1,
    random_state=42
)
stage1_model_ablation.fit(X_train, y_train_stage1)


def compute_metrics(y_true_fine, y_pred_fine, experiment_name, rare_threshold_local=500, debug=None):
    y_true_fine = np.array(y_true_fine)
    y_pred_fine = np.array(y_pred_fine, dtype=object)

    acc = accuracy_score(y_true_fine, y_pred_fine)
    macro_f1 = f1_score(y_true_fine, y_pred_fine, average='macro', zero_division=0)
    weighted_f1 = f1_score(y_true_fine, y_pred_fine, average='weighted', zero_division=0)

    # Robust benign label handling
    benign_fp = np.nan
    benign_candidates = ['Benign', 'BenignTraffic']
    benign_label = next((label for label in benign_candidates if label in np.unique(y_true_fine)), None)
    if benign_label is not None:
        benign_mask = (y_true_fine == benign_label)
        if benign_mask.sum() > 0:
            benign_fp = float(np.mean(y_pred_fine[benign_mask] != benign_label))

    report_dict = classification_report(y_true_fine, y_pred_fine, output_dict=True, zero_division=0)
    support_series = pd.Series(y_true_fine).value_counts()
    rare_classes = support_series[support_series < rare_threshold_local].index.tolist()

    rare_recalls = []
    for cls_name in rare_classes:
        if cls_name in report_dict:
            rare_recalls.append(report_dict[cls_name]['recall'])
    rare_recall_mean = float(np.mean(rare_recalls)) if len(rare_recalls) > 0 else np.nan

    result = {
        'Experiment': experiment_name,
        'Pipeline Accuracy': acc,
        'Pipeline Macro-F1': macro_f1,
        'Pipeline Weighted-F1': weighted_f1,
        'Benign FP Rate': benign_fp,
        f'Rare Recall (<{rare_threshold_local})': rare_recall_mean,
        'Rare Class Count': len(rare_classes)
    }

    if debug is not None:
        result.update(debug)

    return result


def train_stage2_variant(mode='default'):
    train_other_mask_local = (y_train_stage1 == 'Other')
    X2_local = X_train[train_other_mask_local]
    y2_local = df_train.iloc[train_indices[train_other_mask_local]]['label'].values

    if mode == 'default':
        min_samples_local = 3
        k_cap = 5
    elif mode == 'tuned':
        min_samples_local = 2
        k_cap = 3
    else:
        raise ValueError('mode must be default or tuned')
    y2_counts_local = pd.Series(y2_local).value_counts()
    valid_classes_local = y2_counts_local[y2_counts_local >= min_samples_local].index.values

    keep_mask = np.isin(y2_local, valid_classes_local)
    X2_local = X2_local[keep_mask]
    y2_local = y2_local[keep_mask]

    enabled = len(y2_local) > 0 and len(np.unique(y2_local)) >= 2
    if not enabled:
        return None, None, False

    le_local = LabelEncoder()
    y2_enc_local = le_local.fit_transform(y2_local)

    min_class_size_local = pd.Series(y2_local).value_counts().min()
    k_neighbors_local = max(1, min(k_cap, int(min_class_size_local) - 1))

    smote_local = SMOTE(random_state=42, k_neighbors=k_neighbors_local)
    X2_res_local, y2_res_local = smote_local.fit_resample(X2_local, y2_enc_local)

    stage2_local = RandomForestClassifier(
        n_estimators=100,
        max_depth=15,
        class_weight='balanced',
        n_jobs=1,
        random_state=42
    )
    stage2_local.fit(X2_res_local, y2_res_local)

    return stage2_local, le_local, True


def run_pipeline_variant(mode='stage1_only'):
    y_pred_stage1_local = stage1_model_ablation.predict(X_test)
    final_pred_local = y_pred_stage1_local.copy()

    stage1_other_count = int(np.sum(y_pred_stage1_local == 'Other'))
    total_test = int(len(y_pred_stage1_local))

    debug = {
        'Stage1 Other Count': stage1_other_count,
        'Stage2 Enabled': False,
        'Sent To Stage2': 0,
        'Labels Changed By Stage2': 0,
        'Other Routing Rate': stage1_other_count / total_test if total_test > 0 else np.nan
    }

    if mode == 'stage1_only':
        return np.array(final_pred_local, dtype=object), debug

    if mode == 'stage2_default':
        stage2_local, le_local, enabled_local = train_stage2_variant(mode='default')
    elif mode == 'stage2_tuned':
        stage2_local, le_local, enabled_local = train_stage2_variant(mode='tuned')
    else:
        raise ValueError('mode must be stage1_only, stage2_default, or stage2_tuned')

    debug['Stage2 Enabled'] = bool(enabled_local)

    if enabled_local and stage2_local is not None and le_local is not None:
        other_idx_local = np.where(y_pred_stage1_local == 'Other')[0]
        debug['Sent To Stage2'] = int(len(other_idx_local))

        if len(other_idx_local) > 0:
            X_stage2_local = X_test[other_idx_local]
            stage2_pred_enc = stage2_local.predict(X_stage2_local)
            stage2_pred_labels = le_local.inverse_transform(stage2_pred_enc)

            before_labels = final_pred_local[other_idx_local].copy()
            final_pred_local[other_idx_local] = stage2_pred_labels
            debug['Labels Changed By Stage2'] = int(np.sum(before_labels != stage2_pred_labels))

    return np.array(final_pred_local, dtype=object), debug


results_rows = []

pred_stage1_only, dbg_stage1_only = run_pipeline_variant(mode='stage1_only')
results_rows.append(compute_metrics(y_test_fine, pred_stage1_only, 'Stage1 Only', rare_threshold, dbg_stage1_only))

pred_stage2_default, dbg_stage2_default = run_pipeline_variant(mode='stage2_default')
results_rows.append(compute_metrics(y_test_fine, pred_stage2_default, 'Stage1 + Stage2 (Default SMOTE)', rare_threshold, dbg_stage2_default))

pred_stage2_tuned, dbg_stage2_tuned = run_pipeline_variant(mode='stage2_tuned')
results_rows.append(compute_metrics(y_test_fine, pred_stage2_tuned, 'Stage1 + Stage2 (Tuned SMOTE)', rare_threshold, dbg_stage2_tuned))

ablation_df = pd.DataFrame(results_rows)
for col in ablation_df.columns:
    if col != 'Experiment':
        ablation_df[col] = pd.to_numeric(ablation_df[col], errors='coerce')

print('Ablation Results (for paper table):')
display(ablation_df)

print('\nTable output:')
try:
    print(ablation_df.to_markdown(index=False))
except Exception:
    print(ablation_df.to_string(index=False))
    print('\nInstall optional package for markdown export: pip install tabulate')
```
